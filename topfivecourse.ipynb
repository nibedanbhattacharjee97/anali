```python
import pandas as pd

# Load the data
file_path = 'education_data.xlsx'  # Ensure this is the correct path
df = pd.read_excel(file_path)

# 1. Calculate Placement Rate and Average Marks by Course
course_analysis = df.groupby('Course').agg(
    PlacementRate=('PlacementStatus', lambda x: (x == 'Placed').mean() * 100),
    AverageMarks=('Marks', 'mean')
).reset_index()


# 3. Identify Bottom 5 and Top 5 Courses by Placement Rate
bottom_5_courses = course_analysis.nsmallest(5, 'PlacementRate')
top_5_courses = course_analysis.nlargest(5, 'PlacementRate')


# Display the results
print("Bottom 5 Courses (Target for Improvement):")
print(bottom_5_courses[['Course', 'PlacementRate', 'AverageMarks']])
print("\nTop 5 Courses (Best Performing):")
print(top_5_courses[['Course', 'PlacementRate', 'AverageMarks']])


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_5_courses['Course'], top_5_courses['PlacementRate'], color='teal')
plt.title('Top 5 Courses by Placement Rate')
plt.xlabel('Course')
plt.ylabel('Placement Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
